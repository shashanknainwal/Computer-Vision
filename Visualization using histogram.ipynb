```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
img= cv2.imread('/Users/shashank/Desktop/input.jpg')
```

```python
histogram= cv2.calcHist([img],[0],None,[256],[0,256])
```

```python
plt.plot(histogram)
```

```python
color =('b','g','r')
```

```python
for i,col in enumerate(color):
    histogram2= cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histogram2,color=col)
    plt.xlim([0,256])

plt.show()
    
```
