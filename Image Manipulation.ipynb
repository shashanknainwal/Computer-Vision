## Rotation

## Adaptive Threshold

cv2.aptiveThreshold(image, Max Value,Adaptive type, Threshold Type,Block size)
3 types of threshold type
1. ADAPTIVE_THRESH_MEAN_C- based on means of neighbourhood pixels
2. ADAPTIVE_THRESH_GAUSSIAN_C- based on weighted sum of neighbourhood pixels

```python
## Use adaptive type
thresh2= cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,3,5)
```

```python
plt.imshow(thresh2)
plt.show
```

# Edge detection

There are 3 types of edge detection
1. Sobel- vertical and horizontal edges
2. Laplacian- all orientations
3. Canny- Optimal due to low error rate, well defined edges and accurate detection

```python
image= cv2.imread('/Users/shashank/Desktop/Hillary.jpg')
```

```python
image.shape
```

```python
plt.imshow(image)
plt.show
```

```python
# Extract Sobel edges
x= cv2.Sobel(image,cv2.CV_64F,0,1,ksize=5)
y= cv2.Sobel(image,cv2.CV_64F,1,0,ksize=5)
```

```python
plt.imshow(x)
plt.show
```

```python
plt.imshow(y)
plt.show
```

```python
# Extract Laplacian edges
```

```python
laplacian_x= cv2.Laplacian(image,cv2.CV_64F)
```

```python
plt.imshow(laplacian_x)
plt.show
```

```python
## Extract Canny edges
```

```python
canny= cv2.Canny(image,0,170)
```

```python
plt.imshow(canny)
plt.show
```
